```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

players_df = '/content/drive/MyDrive/Trabajo_Final_Coder_Ciencia_1/players_21.csv'
df = pd.read_csv(players_df)
players_df = pd.read_csv(players_df)

# Descripción de Variables

# sofifa_id: Identificador único del jugador en la base de datos de FIFA.
# player_url: URL del perfil del jugador en el sitio de sofifa.
# short_name: Nombre corto del jugador.
# long_name: Nombre completo del jugador.
# age: Edad del jugador.
# dob: Fecha de nacimiento del jugador.
# height_cm: Altura del jugador en centímetros.
# weight_kg: Peso del jugador en kilogramos.
# nationality: Nacionalidad del jugador.
# club_name: Nombre del club al que pertenece el jugador.
# lwb, ldm, cdm, rdm, rwb, lb, lcb, cb, rcb, rb: Habilidades del jugador en diferentes posiciones del campo"""

# siendo estas ultimas:
# LWB (Left Wing Back): Calificación del jugador como lateral izquierdo.
# LDM (Left Defensive Midfielder): Calificación del jugador como mediocampista defensivo izquierdo.
# CDM (Central Defensive Midfielder): Calificación del jugador como mediocampista defensivo central.
# RDM (Right Defensive Midfielder): Calificación del jugador como mediocampista defensivo derecho.
# RWB (Right Wing Back): Calificación del jugador como lateral derecho.
# LB (Left Back): Calificación del jugador como defensor izquierdo.
# LCB (Left Center Back): Calificación del jugador como defensor central izquierdo.
# CB (Center Back): Calificación del jugador como defensor central.
# RCB (Right Center Back): Calificación del jugador como defensor central derecho.
# RB (Right Back): Calificación del jugador como defensor derecho.
```

```python
# Primer grafico con matplotlib, vemos la distribucion de las edades de los jugadores

plt.figure(figsize=(10, 6))
plt.hist(players_df['age'], bins=30, color='blue', edgecolor='black')
plt.title('Distribución de Edades de los Jugadores')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Interpretación:
# La mayoría de los jugadores tienen entre 20 y 30 años, con un pico en los 25 años. Esto sugiere que
# la edad promedio de los jugadores en el dataset está en el rango de 20-30 años.
```

```python
# Segundo grafico con matplotlib, vemos la distribucion de las alturas de los jugadores

plt.figure(figsize=(10, 6))
plt.hist(players_df['height_cm'], bins=30, color='green', edgecolor='black')
plt.title('Distribución de Alturas de los Jugadores')
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Interpretación:
# La altura de los jugadores se concentra en torno a los 175-185 cm, con una distribución normal.
# La mayoría de los jugadores tienen una altura promedio, con pocos jugadores por debajo de 170 cm o por encima de 190 cm.
```

```python
# Tercer grafico con matplotlib, vemos las diferentes nacionalidades de los jugadores

plt.figure(figsize=(12, 8))
nationality_counts = players_df['nationality'].value_counts().head(20)
plt.barh(nationality_counts.index, nationality_counts.values, color='purple')
plt.title('Nacionalidades más Comunes entre los Jugadores')
plt.xlabel('Cantidad de Jugadores')
plt.ylabel('Nacionalidad')
plt.show()

# Interpretación:
# Las nacionalidades más comunes entre los jugadores son Inglaterra, Alemania, España, Francia, Brasil y Argentina
# Esto refleja la popularidad y la fortaleza de estos países en el mundo del fútbol.
```

```python
# Primer grafico con seaborn, vemos la distribucion de los pesos de los jugadores

plt.figure(figsize=(10, 6))
sns.histplot(players_df['weight_kg'], bins=30, kde=True, color='red')
plt.title('Distribución de Pesos de los Jugadores')
plt.xlabel('Peso (kg)')
plt.ylabel('Frecuencia')
plt.show()

# Interpretación:
# La mayoría de los jugadores tienen un peso entre 70 y 80 kg, con una distribución normal.
# Esto sugiere que el peso promedio de los jugadores en el dataset está en este rango.
```

```python
# Segundo grafico con seaborn, vemos una analisis de la relacion entre las edades y las alturas.

plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='height_cm', data=players_df, hue='nationality', palette='viridis', legend=False)
plt.title('Comparación de Edades y Alturas de los Jugadores')
plt.xlabel('Edad')
plt.ylabel('Altura (cm)')
plt.show()

# Interpretación:
# No hay una correlación clara entre la edad y la altura de los jugadores.
# Los puntos dispersos muestran que los jugadores de diferentes edades tienen diversas alturas.
```

```python
# Tercer grafico con seaborn, vemos una analisis de la relacion entre calificacion y altura de los defensores

position_columns_b = ['lwb', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb']

# Asegurarse de que las columnas de posición sean de tipo string y convertirlas a numéricas
for col in position_columns_b:
    players_df[col] = players_df[col].astype(str).str.extract(r'(\d+)').astype(float)

# Crear scatterplots
plt.figure(figsize=(14, 8))

for pos in position_columns_b:
    sns.scatterplot(x='height_cm', y=pos, data=players_df, label=pos.upper())

plt.title('Relación entre Altura y Calificaciones de los defensores')
plt.xlabel('Altura (cm)')
plt.ylabel('Calificación')
plt.legend(title='Posición')
plt.grid(True)
plt.show()

# Interpretación:
# Este gráfico de dispersión muestra la relación entre la altura de los jugadores y sus calificaciones en diferentes
# posiciones de los jugadores que son defensores. Podemos observar cómo la altura de los jugadores se correlaciona con sus calificaciones
# en estas posiciones específicas del campo.
```
